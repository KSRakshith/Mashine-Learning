```python
data=red[(red['residual sugar'] > (np.mean(red['residual sugar']) - 3*np.std(red['residual sugar']))) & (red['residual sugar'] < (np.mean(red['residual sugar']) + 2.15*np.std(red['residual sugar'])))]
sns.boxplot(data)
print(data.shape)
```

```python
data2=data[(data['chlorides'] > (np.mean(data['chlorides']) - 3*np.std(data['chlorides']))) & (data['chlorides'] < (np.mean(data['chlorides']) + 2.7*np.std(data['chlorides'])))]
data2.shape
```

```python
data3=data2[(data2['sulphates'] > (np.mean(data2['sulphates']) - 3*np.std(data2['sulphates']))) & (data2['sulphates'] < (np.mean(data2['sulphates']) + 3*np.std(data2['sulphates'])))]
data3.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
data3_train,data3_test=train_test_split(data3,test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
dt3_train_scale=pd.DataFrame(sc.fit_transform(data3_train),columns=data3.columns)
```

```python
dt3_test_scale=pd.DataFrame(sc.transform(data3_test),columns=data3.columns)
```

```python
print(dt3_train_scale.shape)
print(dt3_test_scale.shape)
```

```python
x_train=dt3_train_scale.drop('pH',axis=1)
x_test=dt3_test_scale.drop('pH',axis=1)
```

```python
y_train=dt3_train_scale.pH
y_test=dt3_test_scale.pH
```

```python
from sklearn.feature_selection import RFE
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
rfe=RFE(lr,9)
```

```python
rfe=rfe.fit(x_train,y_train)
```

```python
list(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

```python
sel_col=x_train.columns[rfe.support_]
```

```python
x_train_rfe=x_train[sel_col]
```

```python
import statsmodels.api as sm
```

```python
x_train_sm=sm.add_constant(x_train_rfe)
```

```python
linmod=sm.OLS(y_train,x_train_sm).fit()
```

```python
linmod.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif['Features']=x_train_rfe.columns
vif['VIF']=[variance_inflation_factor(x_train_rfe.values,i) for i in range(x_train_rfe.shape[1])]
vif['VIF']=round(vif['VIF'],2)
vif=vif.sort_values(by='VIF',ascending=False)
vif
```

```python
x_train_sm.drop('citric acid',axis=1,inplace=True)
```

```python
linmod1=sm.OLS(y_train,x_train_sm).fit()
```

```python
linmod1.summary()
```

```python
vif=pd.DataFrame()
vif['Features']=x_train_sm.columns
vif['VIF']=[variance_inflation_factor(x_train_sm.values,i) for i in range(x_train_sm.shape[1])]
vif['VIF']=round(vif['VIF'],2)
vif=vif.sort_values(by='VIF',ascending=False)
vif
```

```python
x_train_sm.head()
```

```python
y_train_pred_sm=linmod1.predict(x_train_sm)
```

```python
error=y_train-y_train_pred_sm
```

```python
sns.distplot(error)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train,y_train_pred_sm)
```

```python
x_test_sm=x_test_sm[x_train_sm.columns]
```

```python
y_pred_test_sm=linmod1.predict(x_test_sm)
y_pred_train_sm=linmod1.predict(x_train_sm)
```

```python
r2_score(y_test,y_pred_test_sm)
```
