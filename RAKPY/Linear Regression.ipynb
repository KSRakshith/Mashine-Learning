## Training a linear Regression Model

### X and Y array

```python
x=df.drop(['Price','Address'],axis=1)
y=df['Price']
```

```python
x
```

```python
y
```

## Train Test Split

##### it split the data into a training set and testing set.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

## Creating and Training the Model

```python
from sklearn.linear_model import LinearRegression as lr
```

```python
lm=lr()
```

```python
lm.fit(X_train,y_train)
```

## Model Evaluation

Let's evaluate the model by checking out it's coefficients and how we can interpret them

```python
# print the intersept
print(lm.intercept_)
```

```python
lm.coef_
```

```python
coeff_df=pd.DataFrame(lm.coef_,x.columns,columns=['Coefficient'])
coeff_df
```

## Prediction from our model

```python
y_pread=lm.predict(x_test)
```

```python
y_pread_train=lm.predict(x_train)
```

```python
len(y_pread)
```

```python
len(y_test)
```

## Evaluation

```python
from sklearn import metrics
```

```python
metrics.r2_score(y_test,y_pread)
```

```python
metrics.r2_score(y_train,y_pread_train)
```

```python
plt.scatter(y_test,y_pread)
```

## Residual Histogram

```python
sb.distplot((y_test-y_pread),bins=10)
```

```python
sb.distplot((y_train-y_pread_train),bins=10)
```

## Errors

```python
print('MAE: ',metrics.mean_absolute_error(y_test,y_pread))
print('MSE: ',metrics.mean_squared_error(y_test,y_pread))
print('MSLE: ',metrics.mean_squared_log_error(y_test,y_pread))
print('RMsE: ',np.sqrt(metrics.mean_squared_error(y_test,y_pread)))
```
